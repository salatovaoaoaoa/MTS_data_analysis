**Задача №6. Код Цезаря**

Одним из первых в истории примеров шифрования считаются закодированные послания Юлия Цезаря. Римскому полководцу необходимо было послать письменные приказы своим генералам, но он не желал, чтобы в случае чего их прочитали недруги. В результате он стал шифровать свои послания довольно простым методом, которым в последствии стали называть кодом Цезаря.
Идея шифрования была совершенно тривиальной и заключалась в циклическом сдвиге букв на три позиции. В итоге буква А превращалась в D, B – в E, C – в F и так далее. Последние три буквы алфавита переносились на начало. Таким образом, буква Х становилась А, Y – B, а Z – C. Цифры и другие символы не подвергались шифрованию.
Напишите программу, реализующую код Цезаря. Позвольте пользователю ввести фразу и количество символов для сдвига, после чего выведите результирующее сообщение. Убедитесь в том, что ваша программа шифрует как строчные, так и прописные буквы. Так же должна быть возможность указывать отрицательный сдвиг, чтобы можно было использовать вашу программу для расшифровки фраз.

При выполнении задания можно обратить внимание на модуль string.

IN:

Python

2

OUT:

Ravjqp

IN:

Ravjqp

-2

OUT:

Python

````python
from string import ascii_lowercase
from string import ascii_uppercase

def shift_caesar(string, shift):
    
    """Функция реализует шифр Цезаря: циклический сдвиг букв на три позиции"""
    
    lower = ascii_lowercase
    upper = ascii_uppercase

    final_phrase = []

    for char in string:
        if char.islower():
            new_letter = lower[(lower.index(char) + shift) % 26]
        elif char.isupper():
            new_letter = upper[(upper.index(char) + shift) % 26]
        else:
            new_letter = char

        final_phrase.append(new_letter)

    return ''.join(final_phrase)


input_str = input("Введите строку: ")
if not input_str.replace(" ", "").isalpha():
    print("Строка должна содержать только буквы и пробелы")
else:
    try:
        int_str = int(input("Введите сдвиг (целое число): "))
        print(shift_caesar(input_str, int_str))
    except ValueError:
        print("Введите целое число для сдвига")
````

**Задача №7. Вычисляем длину гипотенузы**

Напишите главную функцию, принимающую на вход длины двух катетов прямоугольного треугольника и возвращающую длину гипотенузы, и вспомогательную функцию, рассчитывающую длину гипотенузы по теореме Пифагора. В главной программе должен осуществляется запрос длин сторон у пользователя, вызов вспомогательной функции и вывод на экран полученного результата.

IN:

3

4

OUT:

5

````python
def count_length_hypotenuse(a, b):
    """Функция рассчитывает длину гипотенузы по катетам"""
    return pow(a**2 + b**2, 1/2)

def call_length_hypotenuse():
    
    try:
        cat1 = float(input("Первый катет: "))
        cat2 = float(input("Второй катет: "))
        
        if cat1 <= 0 or cat2 <= 0:
            print("Катеты должны быть положительным числом")
        else:
            print(count_length_hypotenuse(cat1, cat2))
        
    except ValueError:
        print("Введите положительное число") 

call_length_hypotenuse()
````

**Задача №8. Простое число?**

Простое число представляет собой число, больше 1, которое без остатка делится лишь на само себя и на единицу. Напишите функцию для определения того. Является ли введенное число простым. Возвращаемое значение должно быть либо True, либо False. В основной программе, как и ожидается, пользователь должен ввести целое число и получить ответ о том, является ли оно простым.

IN: 23

OUT: True

IN: 4

OUT: False

````python
def chek_determ(num):

    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(pow(num, 1/2)) + 1, 2):
        if num % i == 0:
            return False
    return True

def cheker():

    try:
        numb = int(input("Введите целое положительное число"))
        print(chek_determ(numb))
    except ValueError:
        print("Введите положительное число")

cheker()
````

IN:

**Задача №10. Избавляемся от дубликатов**

В данном упражнении вам предстоит разработать программу, в которой у пользователя будет запрошен список слов, пока он не оставит строку ввода пустой. После этого на экране должны быть показаны слова, введенные пользователем, но без повторов, - каждое по одному разу. При этом слова должны быть отображены в том же порядке, в каком их вводили с клавиатуры. Например. Если пользователь на запрос программы введет следующий список слов:

IN:

first

second

first

third

second

""

OUT:

first

second

third

````python
def words():
    list_words = set()

    while True:
        w = input("Введите слово (или нажмите Enter для завершения): ")
        
        if w.isalpha():
            list_words.add(w)
        else:
            print("Нужно вводить только слова")
            break

        if w == '':
            break

        elif w.isalpha():
            list_words.add(w)
        else:
            print("Нужно вводить только слова")
            break

    if list_words:
        for i in list_words:
            print(i)
    else:
        print("Нет слов... Одни эмоции")

words()
````
